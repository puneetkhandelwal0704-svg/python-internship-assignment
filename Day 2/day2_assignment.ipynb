```python
num1=int(input("enter the first number: "))
num2=int(input("enter the second number: "))
 
n1=num1+num2
print("The sum of the two numbers is: ",n1)

n2=num1-num2
print("The difference of the two numbers is: ",n2)

n3=num1*num2
print("The product of the two numbers is: ",n3)

n4=num1/num2
print("The quotient of the two numbers is: ",n4)

```
