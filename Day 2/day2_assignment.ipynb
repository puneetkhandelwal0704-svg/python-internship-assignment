```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

num1=int(input("enter the first number: ")) #num1=98
num2=int(input("enter the second number: "))#num2=77
 
print("The sum of two numbers is: ",add(num1,num2))
print("the difference of two numbers is:",sub(num1,num2))
print("the product of two numbers is:",mul(num1,num2))
print("the quotient of two numbers is:",div(num1,num2))

```

```python
num=int(input("enter the number: "))
n=str(num)[::-1]
print("The reversed number is: ",n)
if(num==int(n)):
    print("the given number is a palindrome")
else:
    print("the given number is not a palindrome")
 
```
