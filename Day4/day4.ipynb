```python
#Exception handling

try:
    num1 = int(input("Enter first number: "))
    print(10/num1)

except ZeroDivisionError:
    print("An error occurred. Please check your input and try again.")

except ValueError:
    print("An error occurred. Please enter a valid number.")

except TypeError:
    print("An error occurred. Please check your input and try again.")
```

```python
import sqlite3

conn = sqlite3.connect("school.db")

sql = """
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    mobile_no TEXT NOT NULL,
    city TEXT NOT NULL
);
"""

conn.execute(sql)

conn.commit()
conn.close()

print("Table checked/created successfully")
```

```python
import sqlite3

conn = sqlite3.connect("school.db")

conn.execute("""
ALTER TABLE students
ADD COLUMN city TEXT
""")

conn.commit()
conn.close()

print("City column added successfully")
```
