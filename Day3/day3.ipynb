```python
def rec(k):
    if(k>0):
        print(k)
        k=k-1
        rec(k)
    else:
        print(k)
rec(5)

```

```python
with open("myfile.txt","r") as file:
    data=file.readlines()
    for line in data:
        word=line.split()
        print(word)
```
