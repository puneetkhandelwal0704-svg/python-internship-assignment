```python
l=[1,2,3,4,5,4,6,1,3,8]
def remove_duplicates(l): 
    unique_list=[] 
    for i in l:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

print("The list after removing duplicates is:", remove_duplicates(l))
```

```python
num1=int(input("Enter the nunmber:"))

def factorial(num1):
    result=1

    for i in range(1,num1+1):
        result=result*i

    return result

print("Factorial is:", factorial(num1))
```

```python
def is_prime(n):

    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True


num = 7

if is_prime(num):
    print(num, "is Prime")

else:
    print(num, "is Not Prime")
```

```python
def count_case(s):

    upper = 0
    lower = 0

    for i in s:

        if i.isupper():
            upper += 1

        elif i.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)


text = "PyThOn"

count_case(text)
```
